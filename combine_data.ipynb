```python
import pandas as pd
```

```python
Jan = pd.read_csv('data/JC-201801-citibike-tripdata.csv', index_col=0)
Feb = pd.read_csv('data/JC-201802-citibike-tripdata.csv', index_col=0)
Mar = pd.read_csv('data/JC-201803-citibike-tripdata.csv', index_col=0)
Apr = pd.read_csv('data/JC-201804-citibike-tripdata.csv', index_col=0)
May = pd.read_csv('data/JC-201805-citibike-tripdata.csv', index_col=0)
Jun = pd.read_csv('data/JC-201806-citibike-tripdata.csv', index_col=0)
Jul = pd.read_csv('data/JC-201807-citibike-tripdata.csv', index_col=0)
Aug = pd.read_csv('data/JC-201808-citibike-tripdata.csv', index_col=0)
Sept = pd.read_csv('data/JC-201809-citibike-tripdata.csv', index_col=0)
Oct = pd.read_csv('data/JC-201810-citibike-tripdata.csv', index_col=0)
Nov = pd.read_csv('data/JC-201811-citibike-tripdata.csv', index_col=0)
Dec = pd.read_csv('data/JC-201812-citibike-tripdata.csv', index_col=0)
```

```python
combined_tripdata = pd.concat([Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sept, Oct, Nov, Dec])
combined_tripdata
```

```python
combined_tripdata['gender'] = combined_tripdata.gender.astype(str)
combined_tripdata['gender'].dtype
```

```python
combined_tripdata['gender'] = combined_tripdata['gender'].replace("1", 'Male')
combined_tripdata['gender'] = combined_tripdata['gender'].replace("2", 'Female')
combined_tripdata['gender'] = combined_tripdata['gender'].replace("0", 'n/a')
combined_tripdata
```

```python
combined_tripdata.to_csv(r'data/combined_tripdata.csv')
```
